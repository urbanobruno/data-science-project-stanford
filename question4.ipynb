```python
df_final.drop(columns=['SHCODE', 'YEAR', 'WBCTRY'], inplace=True)

df_final_numeric = df_final.drop(columns='COUNTRY').apply(pd.to_numeric, errors='coerce')

threshold = 1_000_000


df_final_numeric = df_final_numeric.apply(lambda x: x.where(x <= threshold, 0))
#df_final_numeric['COUNTRY'] = df_final['COUNTRY']
df_final_numeric.head()
```

```python
df_final_numeric.dropna(inplace=True)
df_final_numeric.head()
```

```python
from sklearn.preprocessing import StandardScaler

df_country = df_final['COUNTRY']
#df_numeric = df_final.drop(columns=['COUNTRY'])
df_final_numeric.replace([np.inf, -np.inf], np.nan, inplace=True)

scaler = StandardScaler()
df_numeric_scaled = pd.DataFrame(scaler.fit_transform(df_final_numeric), columns=df_final_numeric.columns)

df_final_scaled = pd.concat([df_country, df_numeric_scaled], axis=1)
```

```python
df_final_scaled.head()
```

```python
df_aggregated = df_final_scaled.groupby('COUNTRY', as_index=False).mean()
```

```python
df_aggregated.head()
```

```python
df_country = df_aggregated['COUNTRY']
df_numeric = df_aggregated.drop(columns=['COUNTRY'])
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(df_numeric)

explained_variance = pd.DataFrame({
    'Principal Component': [f'PC{i+1}' for i in range(len(pca.explained_variance_ratio_))],
    'Explained Variance Ratio': pca.explained_variance_ratio_,
    'Cumulative Explained Variance': pca.explained_variance_ratio_.cumsum()
})

pca_components = pd.DataFrame(pca.components_, columns=df_numeric.columns, index=[f'PC{i+1}' for i in range(len(pca.components_))])

print("Explained Variance by each Principal Component:")
print(explained_variance)

print("\nPCA Components (loadings):")
print(pca_components)
```

```python
import matplotlib.pyplot as plt

target_variance = 0.8
num_components = np.argmax(explained_variance['Cumulative Explained Variance'] >= target_variance) + 1

plt.figure(figsize=(10, 6))
plt.plot(explained_variance['Principal Component'], explained_variance['Cumulative Explained Variance'], marker='o', linestyle='--', color='b')

plt.axhline(y=target_variance, color='r', linestyle='--')

plt.axvline(x=num_components, color='g', linestyle='--')

plt.text(num_components, target_variance + 0.02, f'{num_components} PCs', ha='center', color='g')
plt.text(len(explained_variance) + 0.5, target_variance, f'{target_variance*100:.0f}% Explained Variance', va='center', color='r')

plt.title('Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)

plt.show()

selected_components = pca_components.iloc[:num_components]

print(f"Number of Principal Components retaining {target_variance*100:.0f}% of the variance: {num_components}")
```

```python
num_components = 12

pca_components_abs = pca_components.iloc[:num_components].abs()

feature_importance = pca_components_abs.sum(axis=0)

top_features = feature_importance.sort_values(ascending=False).index[:num_components]

print("Top 12 features contributing to 80% variance:")
print(top_features)

df_pca = df_numeric[top_features]
```

PCA was done here so that we can identify the top features and draw conclusions only based on them.

```python
df_final_pca = pd.concat([df_country, df_pca], axis=1)
```

```python
df_final_pca.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = df_final_pca.drop('COUNTRY', axis=1)


k_values = range(2, 11)  
inertias = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.plot(k_values, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X = df_final_pca.drop('COUNTRY', axis=1)


k_range = range(2, 11)


silhouette_scores = []


for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {k}, the average silhouette score is : {silhouette_avg}")


plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Coefficient')
plt.title('Silhouette Score vs Number of Clusters')
plt.grid(True)
plt.show()
```

So we can see that the best k is 2. 

```python
optimal_k = 2  
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
cluster_labels = kmeans.fit_predict(X)

df_final_pca['Cluster'] = cluster_labels
```

```python
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X['SHPUPS80'], X['SHPUPS90'], c=cluster_labels, cmap='viridis')
plt.xlabel('SHPUPS80')
plt.ylabel('SHPUPS90')
plt.title('Country Clusters')
plt.colorbar(scatter)
plt.show()
```

```python
feature_pairs = [
    ('SHPUPS85', 'DROP85'),
    ('SHSALP80', 'DROP80'),
    ('N025', 'SHPUPS90'),
    ('GEESEC60', 'SHPUPP60')
]

for feature_x, feature_y in feature_pairs:
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(X[feature_x], X[feature_y], c=cluster_labels, cmap='viridis')
    plt.xlabel(feature_x)
    plt.ylabel(feature_y)
    plt.title(f'{feature_x} vs {feature_y} Clusters')
    plt.colorbar(scatter)
    plt.show()
```
